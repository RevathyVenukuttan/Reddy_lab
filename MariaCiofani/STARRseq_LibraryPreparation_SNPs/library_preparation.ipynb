# Library preparation for all the SNPs

Points to note:
- There are SNPs with point mutations, insertions and deletions. For all of them, the length of the test sequence in the library remains the same = 246


- Oligos will be created for all of the alternate allele and all of them will have 3 phasing, with the SNP in the 1/3rd, 1/2 and 2/3rd position of the oligo


- For SNPs with non-point mutation, if the difference between the reference and the alternate allele sequence is greater than 12, we discard such SNPs as they are dificult to fit in the library

# Sanity checks for the fasta sequences after haplotype mapping

```python
import pandas as pd
import glob

def fasta_to_df(file,file_name,out_dir):
    myDict = {}
    with open(file, 'rU') as fileData:
        tag = None
        for line in fileData:
            line = line.strip()
            if not line:
                continue
            if line[0] == '>':
                tag = line[1:]
                myDict[tag] = ''
            else:
                assert tag is not None, 'Invalid format, found gene without tag'
                myDict[tag] += line
    
    ### create df from the dict and wrangle it to get columns in the order:
    ### 'snp_id','context','ref','count','coordinates','position','ancestry','method','1000G','sequence'
    
    df = pd.DataFrame.from_dict(myDict, orient='index').reset_index()
    df = df.rename(columns={'index':'header',0:'sequence'})
    df[['snp_id','context','ref','count','coordinates','position','ancestry','method','1000G']] = df.header.str.split('/', expand=True)
    for name in ['context','ref','count','coordinates','position','ancestry','method','1000G']:
        df[name]=df[name].apply(lambda x: x.split('=')[1])
    
    df['snp_id'] = df['snp_id'].apply(lambda x: x.split(':')[0])
    df = df[['snp_id','context','ref','count','coordinates','position','ancestry','method','1000G','sequence']]
    df.to_csv('%s/%s.txt'%(out_dir,file_name), sep='\t', index=False)
    return df
```
