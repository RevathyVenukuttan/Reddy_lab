```python
%%bash
cd /data/reddylab/Revathy/collabs/Kari/ENCODE/data/corrected_peaks
cat A001-K562-rep1.f3q10.sorted.dups_marked.macs_keep_dups_norm_peaks.narrowPeak \
A001-K562-rep2.f3q10.sorted.dups_marked.macs_norm_peaks.narrowPeak \
A001-K562-rep3.f3q10.sorted.dups_marked.macs_norm_peaks.narrowPeak \
> /data/reddylab/Revathy/collabs/Kari/ENCODE/data/corrected_peaks/K562_wgSTARR_MACS_peaks.bed
```

```python
%%bash
module load bedtools2
cat /data/reddylab/Revathy/collabs/Kari/ENCODE/data/corrected_peaks/K562_wgSTARR_MACS_peaks.bed \
| sort -k1,1 -k2,2n \
> /data/reddylab/Revathy/collabs/Kari/ENCODE/data/corrected_peaks/K562_wgSTARR_MACS_peaks_sort.bed
```

```python
my_cols = ['chr','start','end','name','score','strand','signalVal','pval','qval','peak']

test = pd.read_csv('/data/reddylab/Revathy/collabs/Kari/ENCODE/data/corrected_peaks/K562_wgSTARR_MACS_peaks_sort.bed',
                   sep="\s+|;|:", names=my_cols, engine='python')

test.to_csv('/data/reddylab/Revathy/collabs/Kari/ENCODE/data/corrected_peaks/K562_wgSTARR_MACS_peaks_sort_corrected.bed', sep='\t', index=False, header=None)
```

```python
%%bash
module load bedtools2
bedtools merge -i /data/reddylab/Revathy/collabs/Kari/ENCODE/data/corrected_peaks/K562_wgSTARR_MACS_peaks_sort_corrected.bed \
> /data/reddylab/Revathy/collabs/Kari/ENCODE/data/corrected_peaks/K562_wgSTARR_MACS_UnionPeakSet.txt
```
