```python
%%writefile /data/reddylab/Revathy/collabs/Jamborees/fcc_capstone/plot_summaryHeatmap.V2.py
import pandas as pd
import numpy as np
from numpy import trapz
from deeptools import heatmapper 
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_context("talk")
sns.set_style("whitegrid")
sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})

hm=heatmapper.heatmapper()
hm.read_matrix_file('/data/reddylab/Revathy/collabs/Jamborees/fcc_capstone/data/enhancer_call_comparison/10032022/ASTARR_Lenti_wgSTARR.promoter_vs_distal.dustin_seq_feature.V2.refpoint.mat.gz')


# ChIP_ASTARR=[]
# ChIP_Lenti=[]
# ChIP_wgSTARR=[]

AUC_ASTARRp=[]
AUC_ASTARRd=[]
AUC_Lentip=[]
AUC_Lentid=[]
AUC_wgSTARRp=[]
AUC_wgSTARRd=[]

label=hm.__dict__['parameters']['sample_labels']

for ii in range(len(label)):
    
    max_val_ASTARRp=np.max(hm.matrix.matrix[:5000, ((ii)*400)+150:((ii+1)*400)-150], axis=0)
    pseudo_count_ASTARRp=np.percentile(max_val_ASTARRp,10)
    logFC_ASTARRp=np.log2((np.nanmean(hm.matrix.matrix[:5000, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_ASTARRp, axis=0))/
                       (np.nanmean(hm.matrix.matrix[5000:10000, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_ASTARRp, axis=0)))
    AUC_ASTARRp.append(trapz(logFC_ASTARRp, dx=10))


    max_val_Lentip=np.max(hm.matrix.matrix[10000:15000, ((ii)*400)+150:((ii+1)*400)-150], axis=0)
    pseudo_count_Lentip=np.percentile(max_val_Lentip,10)
    logFC_Lentip=np.log2((np.nanmean(hm.matrix.matrix[10000:15000, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_Lentip, axis=0))/
                       (np.nanmean(hm.matrix.matrix[15000:19999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_Lentip, axis=0)))
    AUC_Lentip.append(trapz(logFC_Lentip, dx=10))


    max_val_wgSTARRp=np.max(hm.matrix.matrix[19999:24999, ((ii)*400)+150:((ii+1)*400)-150], axis=0)
    pseudo_count_wgSTARRp=np.percentile(max_val_wgSTARRp,10)
    logFC_wgSTARRp=np.log2((np.nanmean(hm.matrix.matrix[19999:24999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_wgSTARRp, axis=0))/
                       (np.nanmean(hm.matrix.matrix[24999:29999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_wgSTARRp, axis=0)))
    AUC_wgSTARRp.append(trapz(logFC_wgSTARRp, dx=10))

    
    max_val_ASTARRd=np.max(hm.matrix.matrix[29999:34999, ((ii)*400)+150:((ii+1)*400)-150], axis=0)
    pseudo_count_ASTARRd=np.percentile(max_val_ASTARRd,10)
    logFC_ASTARRd=np.log2((np.nanmean(hm.matrix.matrix[29999:34999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_ASTARRd, axis=0))/
                       (np.nanmean(hm.matrix.matrix[34999:39999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_ASTARRd, axis=0)))
    AUC_ASTARRd.append(trapz(logFC_ASTARRd, dx=10))

        
    max_val_Lentid=np.max(hm.matrix.matrix[39999:44999, ((ii)*400)+150:((ii+1)*400)-150], axis=0)
    pseudo_count_Lentid=np.percentile(max_val_Lentid,10)
    logFC_Lentid=np.log2((np.nanmean(hm.matrix.matrix[39999:44999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_Lentid, axis=0))/
                       (np.nanmean(hm.matrix.matrix[44999:49999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_Lentid, axis=0)))
    AUC_Lentid.append(trapz(logFC_Lentid, dx=10))

    
    max_val_wgSTARRd=np.nanmax(hm.matrix.matrix[49999:54999, ((ii)*400)+150:((ii+1)*400)-150], axis=0)
    pseudo_count_wgSTARRd=np.percentile(max_val_wgSTARRd,10)
    logFC_wgSTARRd=np.log2((np.nanmean(hm.matrix.matrix[49999:54999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_wgSTARRd, axis=0))/
                       (np.nanmean(hm.matrix.matrix[54999:, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_wgSTARRd, axis=0)))
    AUC_wgSTARRd.append(trapz(logFC_wgSTARRd, dx=10))


DF_ASTARR = pd.DataFrame(np.column_stack([label, AUC_ASTARRp, AUC_ASTARRd]), columns=['Label', 'ASTARR_Promoter', 'ASTARR_Distal'])
DF_Lenti = pd.DataFrame(np.column_stack([label, AUC_Lentip, AUC_Lentid]), columns=['Label', 'Lenti_Promoter', 'Lenti_Distal'])
DF_wgSTARR = pd.DataFrame(np.column_stack([label, AUC_wgSTARRp, AUC_wgSTARRd]), columns=['Label', 'wgSTARR_Promoter', 'wgSTARR_Distal'])

DF_ASTARR[['ASTARR_Promoter','ASTARR_Distal']]=DF_ASTARR[['ASTARR_Promoter','ASTARR_Distal']].apply(pd.to_numeric, errors='coerce', axis=1)
DF_Lenti[['Lenti_Promoter','Lenti_Distal']]=DF_Lenti[['Lenti_Promoter','Lenti_Distal']].apply(pd.to_numeric, errors='coerce', axis=1) 
DF_wgSTARR[['wgSTARR_Promoter','wgSTARR_Distal']]=DF_wgSTARR[['wgSTARR_Promoter','wgSTARR_Distal']].apply(pd.to_numeric, errors='coerce', axis=1)

DF = DF_ASTARR.merge(DF_Lenti, on='Label', how='inner').merge(DF_wgSTARR, on='Label', how='inner')

DF = DF.set_index('Label')
DF = DF.fillna(0)
sns.clustermap(DF.loc[:, ['ASTARR_Promoter','Lenti_Promoter','wgSTARR_Promoter','ASTARR_Distal','Lenti_Distal','wgSTARR_Distal']],
               method='median',
               figsize=[8,10],
               vmin=10,
               vmax=80,
               cmap='magma',
               col_cluster= False, 
               cbar_pos=(0.02, 0.8, 0.05, 0.18))

DF.reset_index().to_csv('data/enhancer_call_comparison/bubble_plots/ASTARR_Lenti_wgSTARR.Promoter_Distal.AUC_summary_heatmaps.signal_table.dustin_seq_feature.V2.txt', sep='\t', index=False)
plt.savefig('data/enhancer_call_comparison/bubble_plots/ASTARR_Lenti_wgSTARR.Promoter_Distal.AUC_summary_heatmaps.dustin_seq_feature.V2.png')
```

```python
%%writefile /data/reddylab/Revathy/collabs/Jamborees/fcc_capstone/plot_summaryHeatmap.V3.py
import pandas as pd
import numpy as np
from numpy import trapz
from deeptools import heatmapper 
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_context("talk")
sns.set_style("whitegrid")
sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})

hm=heatmapper.heatmapper()
hm.read_matrix_file('/data/reddylab/Revathy/collabs/Jamborees/fcc_capstone/data/enhancer_call_comparison/10032022/ASTARR_Lenti_wgSTARR.promoter_vs_distal.dustin_seq_feature.V3_nfe.refpoint.mat.gz')


# ChIP_ASTARR=[]
# ChIP_Lenti=[]
# ChIP_wgSTARR=[]

AUC_ASTARRp=[]
AUC_ASTARRd=[]
AUC_Lentip=[]
AUC_Lentid=[]
AUC_wgSTARRp=[]
AUC_wgSTARRd=[]

label=hm.__dict__['parameters']['sample_labels']

for ii in range(len(label)):
    
    max_val_ASTARRp=np.max(hm.matrix.matrix[:5000, ((ii)*400)+150:((ii+1)*400)-150], axis=0)
    pseudo_count_ASTARRp=np.percentile(max_val_ASTARRp,10)
    logFC_ASTARRp=np.log2((np.nanmean(hm.matrix.matrix[:5000, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_ASTARRp, axis=0))/
                       (np.nanmean(hm.matrix.matrix[5000:10000, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_ASTARRp, axis=0)))
    AUC_ASTARRp.append(trapz(logFC_ASTARRp, dx=10))


    max_val_Lentip=np.max(hm.matrix.matrix[10000:15000, ((ii)*400)+150:((ii+1)*400)-150], axis=0)
    pseudo_count_Lentip=np.percentile(max_val_Lentip,10)
    logFC_Lentip=np.log2((np.nanmean(hm.matrix.matrix[10000:15000, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_Lentip, axis=0))/
                       (np.nanmean(hm.matrix.matrix[15000:19999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_Lentip, axis=0)))
    AUC_Lentip.append(trapz(logFC_Lentip, dx=10))


    max_val_wgSTARRp=np.max(hm.matrix.matrix[19999:24999, ((ii)*400)+150:((ii+1)*400)-150], axis=0)
    pseudo_count_wgSTARRp=np.percentile(max_val_wgSTARRp,10)
    logFC_wgSTARRp=np.log2((np.nanmean(hm.matrix.matrix[19999:24999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_wgSTARRp, axis=0))/
                       (np.nanmean(hm.matrix.matrix[24999:29999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_wgSTARRp, axis=0)))
    AUC_wgSTARRp.append(trapz(logFC_wgSTARRp, dx=10))

    
    max_val_ASTARRd=np.max(hm.matrix.matrix[29999:34999, ((ii)*400)+150:((ii+1)*400)-150], axis=0)
    pseudo_count_ASTARRd=np.percentile(max_val_ASTARRd,10)
    logFC_ASTARRd=np.log2((np.nanmean(hm.matrix.matrix[29999:34999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_ASTARRd, axis=0))/
                       (np.nanmean(hm.matrix.matrix[34999:39999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_ASTARRd, axis=0)))
    AUC_ASTARRd.append(trapz(logFC_ASTARRd, dx=10))

        
    max_val_Lentid=np.max(hm.matrix.matrix[39999:44999, ((ii)*400)+150:((ii+1)*400)-150], axis=0)
    pseudo_count_Lentid=np.percentile(max_val_Lentid,10)
    logFC_Lentid=np.log2((np.nanmean(hm.matrix.matrix[39999:44999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_Lentid, axis=0))/
                       (np.nanmean(hm.matrix.matrix[44999:49999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_Lentid, axis=0)))
    AUC_Lentid.append(trapz(logFC_Lentid, dx=10))

    
    max_val_wgSTARRd=np.nanmax(hm.matrix.matrix[49999:54999, ((ii)*400)+150:((ii+1)*400)-150], axis=0)
    pseudo_count_wgSTARRd=np.percentile(max_val_wgSTARRd,10)
    logFC_wgSTARRd=np.log2((np.nanmean(hm.matrix.matrix[49999:54999, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_wgSTARRd, axis=0))/
                       (np.nanmean(hm.matrix.matrix[54999:, ((ii)*400)+150:((ii+1)*400)-150]+pseudo_count_wgSTARRd, axis=0)))
    AUC_wgSTARRd.append(trapz(logFC_wgSTARRd, dx=10))


DF_ASTARR = pd.DataFrame(np.column_stack([label, AUC_ASTARRp, AUC_ASTARRd]), columns=['Label', 'ASTARR_Promoter', 'ASTARR_Distal'])
DF_Lenti = pd.DataFrame(np.column_stack([label, AUC_Lentip, AUC_Lentid]), columns=['Label', 'Lenti_Promoter', 'Lenti_Distal'])
DF_wgSTARR = pd.DataFrame(np.column_stack([label, AUC_wgSTARRp, AUC_wgSTARRd]), columns=['Label', 'wgSTARR_Promoter', 'wgSTARR_Distal'])

DF_ASTARR[['ASTARR_Promoter','ASTARR_Distal']]=DF_ASTARR[['ASTARR_Promoter','ASTARR_Distal']].apply(pd.to_numeric, errors='coerce', axis=1)
DF_Lenti[['Lenti_Promoter','Lenti_Distal']]=DF_Lenti[['Lenti_Promoter','Lenti_Distal']].apply(pd.to_numeric, errors='coerce', axis=1) 
DF_wgSTARR[['wgSTARR_Promoter','wgSTARR_Distal']]=DF_wgSTARR[['wgSTARR_Promoter','wgSTARR_Distal']].apply(pd.to_numeric, errors='coerce', axis=1)

DF = DF_ASTARR.merge(DF_Lenti, on='Label', how='inner').merge(DF_wgSTARR, on='Label', how='inner')

DF = DF.set_index('Label')
DF = DF.fillna(0)
sns.clustermap(DF.loc[:, ['ASTARR_Promoter','Lenti_Promoter','wgSTARR_Promoter','ASTARR_Distal','Lenti_Distal','wgSTARR_Distal']],
               method='median',
               figsize=[8,10],
               vmin=10,
               vmax=80,
               cmap='magma',
               col_cluster= False, 
               cbar_pos=(0.02, 0.8, 0.05, 0.18))

DF.reset_index().to_csv('data/enhancer_call_comparison/bubble_plots/ASTARR_Lenti_wgSTARR.Promoter_Distal.AUC_summary_heatmaps.signal_table.dustin_seq_feature.V3_nfe.txt', sep='\t', index=False)
plt.savefig('data/enhancer_call_comparison/bubble_plots/ASTARR_Lenti_wgSTARR.Promoter_Distal.AUC_summary_heatmaps.dustin_seq_feature.V3_nfe.png')
```

```python
import warnings
import pandas as pd
from pandas.core.common import SettingWithCopyWarning
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)

### Sequence_model_scores ###

seq_feature = pd.read_csv('/data/reddylab/Revathy/collabs/Jamborees/fcc_capstone/data/Dustin_seq_features/Fig3_motifs_Z.txt', sep='\t')
seq_feature.rename(columns={'Unnamed: 0':'Label'}, inplace=True)

seq_feature['Label'][0] = 'JUN'
seq_feature['Label'][1] = 'NFE2'
seq_feature['Label'][2] = 'GATA1'
seq_feature['Label'][4] = 'ELK1'
seq_feature['Label'][5] = 'SP1'
seq_feature['Label'][11] = 'KLF10'
seq_feature['Label'][12] = 'NFYa'
seq_feature['Label'][13] = 'PU1'
seq_feature['Label'][14] = 'STAT1'
seq_feature['Label'][16] = 'ATF1'

for col in ['ASTARR_Promoter', 'ASTARR_Distal', 'Lenti_Promoter','Lenti_Distal', 'wgSTARR_Promoter', 'wgSTARR_Distal']:
    seq_feature[col]=(seq_feature[col]-min(seq_feature[col]))/(max(seq_feature[col])-min(seq_feature[col]))

    
### Genomic_context_scores ###

gen_context = pd.read_csv('/data/reddylab/Revathy/collabs/Jamborees/fcc_capstone/data/enhancer_call_comparison/bubble_plots/ASTARR_Lenti_wgSTARR.Promoter_Distal.AUC_summary_heatmaps.signal_table.dustin_seq_feature.V2.txt', sep='\t')

for col in ['ASTARR_Promoter', 'ASTARR_Distal', 'Lenti_Promoter','Lenti_Distal', 'wgSTARR_Promoter', 'wgSTARR_Distal']:
    gen_context[col]=(gen_context[col]-min(gen_context[col]))/(max(gen_context[col])-min(gen_context[col]))

    
### merged ### 

merged = gen_context.merge(seq_feature, on='Label', how='inner')
merged
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['pdf.fonttype'] = 42

sns.set_context("talk")
sns.set_style("whitegrid")
sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})

assays=['ASTARR','Lenti','wgSTARR']
rho = "\u03C1"

for assay in assays:
    x_prom='%s_Promoter_x'%assay
    y_prom='%s_Promoter_y'%assay
    
    fig, ax = plt.subplots(figsize=[8,8])
    sns.scatterplot(data=merged, x=x_prom, y=y_prom, hue='Label', palette='tab20')
    for x, y, t in zip(
    merged[x_prom],
    merged[y_prom], 
    merged['Label']):
        ax.annotate('{}'.format(t), xy=(x, y), xytext=(5, 0), ha='left', textcoords='offset points', fontsize=13)

#     ax.annotate('{}'.format(t), xy=(x, y), xytext=(5, 0), ha='left', textcoords='offset points', fontsize=10)
    Sp_p = merged[x_prom].corr(merged[y_prom], method='spearman')
    Pear_p = merged[x_prom].corr(merged[y_prom], method='pearson')
    ax.legend([],[], frameon=False)
#     lgnd = plt.legend(bbox_to_anchor=(1.25, 1.05), frameon=False, markerscale=1)
    plt.xlabel('Genomic context')
    plt.ylabel('Sequence models')
    plt.title(r'%s Promoter, Sp. ${\rho}$ = %.2f, Pr. ${\rho}$ = %.2f'%(assay, Sp_p, Pear_p))
    sns.despine()
    plt.show()
    fig.tight_layout()
#     fig.savefig('/data/reddylab/Revathy/collabs/Jamborees/fcc_capstone/data/enhancer_call_comparison/scatterplots/%s.Promoter.Seq_vs_Genomic.pdf'%assay)
    
    
    x_dist='%s_Distal_x'%assay
    y_dist='%s_Distal_y'%assay
    
    fig, ax = plt.subplots(figsize=[8,8])
    sns.scatterplot(data=merged, x=x_dist, y=y_dist, hue='Label', palette='tab20')
    for x, y, t in zip(merged[x_dist], merged[y_dist], merged['Label']):
        ax.annotate('{}'.format(t), xy=(x, y), xytext=(5, 0), ha='left', textcoords='offset points', fontsize=13)
    ax.legend([],[], frameon=False)
    Sp_d = merged[x_dist].corr(merged[y_dist], method='spearman')
    Pear_d = merged[x_dist].corr(merged[y_dist], method='pearson')
    
#      lgnd = plt.legend(bbox_to_anchor=(1.25, 1.05), frameon=False, markerscale=1)
    plt.xlabel('Genomic context')
    plt.ylabel('Sequence models')
    plt.title(r'%s Distal, Sp. ${\rho}$ = %.2f, Pr. ${\rho}$ = %.2f'%(assay, Sp_d, Pear_d))
    sns.despine()
    plt.show()
    fig.tight_layout()
#     fig.savefig('/data/reddylab/Revathy/collabs/Jamborees/fcc_capstone/data/enhancer_call_comparison/scatterplots/%s.Distal.Seq_vs_Genomic.pdf'%assay)
    
#     lgnd = plt.legend(bbox_to_anchor=(1.5, 1), frameon=False, markerscale=1)   
    
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import patches as mpatches
import seaborn as sns
from palettable.cartocolors.qualitative import Bold_10, Vivid_10_r
from matplotlib import markers
%matplotlib inline

from scipy.stats import pearsonr, spearmanr

def corrfunc(x,y, ax=None, **kws):
    """Plot the correlation coefficient in the top left hand corner of a plot."""
    r, _ = pearsonr(x, y)
    ax = ax or plt.gca()
    # Unicode for lowercase rho (ρ)
    rho = "\u03C1" # u'\u03C1'
    ax.annotate('%s = %.2f' % (rho, r), xy=(0.9,1), xycoords=ax.transAxes)

sns.set_context("talk")
sns.set_style("whitegrid")
sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})

fig, ax = plt.subplots(3, 2, figsize=(9,12), sharex=True, sharey=True)
labels=['ASTARR_Promoter', 'ASTARR_Distal', 'Lenti_Promoter','Lenti_Distal', 'wgSTARR_Promoter', 'wgSTARR_Distal']

for i,col in enumerate(labels):  

    axi = ax.flatten()[i]
    axi.scatter(gen_context[col],seq_feature[col], marker='o', s=50, 
                c=(Bold_10.hex_colors+Vivid_10_r.hex_colors)[:17])
    cor = gen_context[col].corr(seq_feature[col])
    
    axi.set_xlim(axi.get_xlim())
    axi.set_ylim(axi.get_ylim())
    axi.set_title('%s, Corr = %.2f'%(col, cor))
    fig.tight_layout()
     
        
# for i in range(1,len(seq_feature['Label'])):
#     lgnd = plt.legend(seq_feature['Label'][i], bbox_to_anchor=(1.5, 1), frameon=False, markerscale=2)

fig.text(0.5, 0.004, 'Genomic Context', ha='center')
fig.text(0.004, 0.5, 'Sequence Features', va='center', rotation='vertical')
fig.subplots_adjust(top=0.95)
sns.despine()
plt.show()
```
