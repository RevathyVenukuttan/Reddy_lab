### Subset of inactive peaks that follow the same distribution as the input peaks

```python
import random 
import numpy as np
import pandas as pd

random.seed(42)

count_inactive, bin_edge = np.histogram(inactive_peaks['log2FC'], bins = bins_inactive)
# count_input, bin_edge = np.histogram(input_peaks['log2FC'], bins = bins_input)

def random_sample(count_df1, count_df2, bin_size):
    m = [0]
    if (count_df1[bin_size]>=int(count_df2[bin_size]) and (count_df2[bin_size]!=0)):
        m = list(random.sample(range(0, count_df1[bin_size]), int(count_df2[bin_size])))
        
    elif (count_df1[bin_size]<int(count_df2[bin_size]) and (count_df1[bin_size]!=0)):
        range_list = list(range(0, count_df1[bin_size]))
        m = list(np.random.choice(range_list, size=int(count_df2[bin_size])))
        
    elif (count_df1[bin_size]==int(count_df2[bin_size]) and (count_df1[bin_size]!=0)):
        m = list(range(0, count_df1[bin_size]))
    
    return m 


df=None
for i in range(0,100):
    bin_start = bins_inactive[i]
    bin_stop = bins_inactive[i+1]
    inactive_v1 = inactive_peaks[(inactive_peaks['log2FC']>=bin_start)&(inactive_peaks['log2FC']<bin_stop)].reset_index()
    inactive_v1.drop(columns='index', inplace=True)

    if not inactive_v1.empty:
        sample_size = random_sample(count_inactive, count_input, i)
        inactive_random = inactive_v1.loc[sample_size]
        if df is None:
            df = inactive_random
        else:
            df = pd.concat([df,inactive_random])  


    
plt.figure(figsize=(10,8))
plt.hist(df['log2FC'], density=True, bins=500, alpha=0.5, label="inactive_peaks")
plt.hist(input_peaks['log2FC'], density=True, bins=500, alpha=0.5, label="input_peaks")
plt.xlabel("Data", size=14)
plt.ylabel("Density", size=14)
plt.legend(loc='upper right')
plt.xlim(-10,15)
plt.ylim(0,1.2)
sns.despine()
plt.show()

```

```python
import random 
import numpy as np
import pandas as pd

random.seed(42)
df=None

count_active, bin_edge_active = np.histogram(all_active_peaks['5_pct_gc'], bins = 100)

sample_size = ((count_active.astype(int)/sum(count_active))*5930).astype(int)
for i in range(0,100):
    bin_start = bin_edge_active[i]
    bin_stop = bin_edge_active[i+1]
    all_inactive_peaks_v1 = all_inactive_peaks[(all_inactive_peaks['8_pct_gc']>=bin_start)&(all_inactive_peaks['8_pct_gc']<bin_stop)].reset_index()
    all_inactive_peaks_v1.drop(columns='index', inplace=True)
    
    count_active = count_active.astype(int)
    m = list(range(0, sample_size[i]))
    
    if not all_inactive_peaks_v1.empty:
        all_inactive_peaks_random = all_inactive_peaks_v1.loc[m]
        
        if df is None:
            df = all_inactive_peaks_random
        else:
            df = pd.concat([df,all_inactive_peaks_random])  

# df[['chr','start','end']] = df['peak'].str.split('_', expand=True)
df = df.dropna().astype({'2_usercol':'int', '3_usercol':'int'})
df[['#1_usercol','2_usercol','3_usercol']].dropna().to_csv('/data/reddylab/Revathy/collabs/Jamborees/fcc_capstone/data/inputs/intersections/ASTARR_Lenti_wgSTARR_inactive_with_input.threshold0.5.coordinates_corrected.GC_content.random5000.bed', sep='\t', index=False, header=None)
# df.to_csv('/data/reddylab/Revathy/collabs/Maria/human-th-ms_new/data_v1/ATACseq/raw_data/peak_calling/h1_hESC_rep1_rep2_random_subsample_on_Th_distribution.ATAC_peaks.txt', sep='\t', index=False)

plt.figure(figsize=(10,8))
plt.hist(df['8_pct_gc'], bins=100, alpha=0.5, label="all_inactive_peaks")
plt.hist(all_active_peaks['5_pct_gc'], bins=100, alpha=0.5, label="all_active_peaks")
plt.xlabel("Data", size=14)
plt.ylabel("Count", size=14)
plt.legend(loc='upper right')
sns.despine()

plt.figure(figsize=(10,8))
plt.hist(df['8_pct_gc'], density=True, bins=100, alpha=0.5, label="all_inactive_peaks")
plt.hist(all_active_peaks['5_pct_gc'], density=True, bins=100, alpha=0.5, label="all_active_peaks")
plt.xlabel("Data", size=14)
plt.ylabel("Density", size=14)
plt.legend(loc='upper right')
sns.despine()
plt.show()

```
