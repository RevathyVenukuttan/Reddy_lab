```python
import pyBigWig
import pandas as pd
import time

start_time = time.time()

bw = pyBigWig.open("/data/reddylab/gjohnson/whole_genome_STARRseq/wgss3/peak_calls/cradle/effect_of_mi/corrected_bigwigs/900/Dex_mean_corrected.bw")
bed = open("/data/reddylab/gjohnson/whole_genome_STARRseq/wgss3/peak_calls/cradle/callPeak_against_DMSO/Dex_vs_DMSO/Dex_vs_DMSO_induced_peaks.bed")


bed_file = pd.read_csv('/data/reddylab/gjohnson/whole_genome_STARRseq/wgss3/peak_calls/cradle/callPeak_against_DMSO/Dex_vs_DMSO/Dex_vs_DMSO_induced_peaks.bed', sep='\t',
                       names=['chr','start','end','chr:start-end','4','5','6','7','8','9','10','11','12'])

bed_file = bed_file[['chr','start','end','chr:start-end']]


seq_list = []
for n in range(len(bed_file)):
    chr = bed_file['chr'][n]
    start = bed_file['start'][n]
    end = bed_file['end'][n]
    ID = bed_file['chr:start-end'][n]
    diff = end-start
    for i in range(diff):
        if (start!=end):
            new_end = start+1
            seq = (chr, start, new_end, ID)
            seq_list.append(seq)
            start = start+1
            i = i+1

expanded_bed_df = pd.DataFrame(seq_list, columns=['chr','start','end','chr:start-end'])

value=[]
for line in bed:
    cols = line.strip().split()
    value.append(bw.values(cols[0], int(cols[1]), int(cols[2])))

value_list  = [item for sublist in value for item in sublist]

expanded_bed_df['value'] = value_list
idx = expanded_bed_df.groupby(['chr:start-end'])['value'].transform(max) == expanded_bed_df['value']
expanded_bed = expanded_bed_df[idx]
expanded_bed_subset = expanded_bed[['chr','start','end']]
expanded_bed_subset.to_csv('/data/reddylab/Revathy/collabs/Graham/script/Dex_vs_DMSO_induced_peaks_with_max_bw_signal.bed', sep='\t', header=None, index=False)

print("Done!")
print("Elapsed: {0} minutes {1:0f} seconds".format((time.time() - start_time) // 60, (time.time() - start_time) % 60))
```
